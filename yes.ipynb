```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("ecommerce_dataset_updated.csv")

df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Data Cleaning

df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Purchase_Date"] = pd.to_datetime(
    df["Purchase_Date"],
    dayfirst=True
)
```

```python
df.head()
```

```python
df["Month"] = df["Purchase_Date"].dt.month_name()

df["Year"] = df["Purchase_Date"].dt.year
```

```python
df
```

```python
df["Final_Price(Rs.)"].sum()
```

```python
df["Final_Price(Rs.)"].mean()
```

```python
df["Final_Price(Rs.)"].max()
```

```python
df["Final_Price(Rs.)"].min()
```

```python
category_sales = df.groupby("Category")["Final_Price(Rs.)"].sum()

print(category_sales)
```

```python
payment = df["Payment_Method"].value_counts()

print(payment)
```

```python
monthly = df.groupby("Month")["Final_Price(Rs.)"].sum()

print(monthly)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=category_sales.index,
    y=category_sales.values
)

plt.title("Revenue by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.pie(
    payment,
    labels=payment.index,
    autopct="%1.1f%%"
)

plt.title("Payment Methods")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    x=monthly.index,
    y=monthly.values,
    marker="o"
)

plt.title("Monthly Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Discount (%)",
    y="Final_Price(Rs.)"
)

plt.title("Discount vs Final Price")
plt.show()
```
